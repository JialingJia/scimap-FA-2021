# Mentor Tree Category

### Skewness

Tree is tilted by mentee's gender. If mentee's gender is male, the branch will bend to the left while female' branch bend to the right. Therefore, the tree skewness measures the asymmetry of a tree's shape.

#### Askew: the mentortree is tilted due to uneven gender distribution

```python
from IPython.display import SVG, display
display(SVG('thin.svg')) 
```

#### Balanced: the mentortree is not polarized by gender difference

```python
display(SVG('balanced.svg')) 
```

### Growth/Height

A tree growth represents its height, referring to how many generations of mentorship exist in a mentortree. A low mentortree usually means that there are few mentee generations while a tall mentortree grows from one generation to another generation.

#### Low: the mentortree only has few generations

```python
display(SVG('low.svg')) 
```

#### Tall: the mentortree has lots of generations

```python
display(SVG('high.svg')) 
```

### Strength

Since mentors may have few or many mentees, a tree strength represents how many mentees the mentor has over different generations. A thin mentortree only has single or few mentees from one generation to another generation while a thick mentortree has lots of mentees over certain generations

#### Thin: the mentortree only has few mentees over different generations

```python
display(SVG('thin.svg')) 
```

#### Thick: the mentortree has many mentees in specific generations

```python
display(SVG('thick.svg')) 
```

### Canopy density

Trees with dense canopies cast more shades. Those with thin canopies will have more light penetrations. Therefore, the canopy density of a mentortree is a combined effect of mentee's gender distribution, the amount of mentees and generations.

#### Scarce: the mentortree is naked with few single-gender mentees.

```python
display(SVG('thin.svg')) 
```

#### Scattered: the mentortree is scattered over different generations. A larger gender difference appears in the early generations

```python
display(SVG('scattered.svg')) 
```

#### Crowded: the mentortree is crowded with mentees continuously over different generations.

```python
display(SVG('crowded.svg')) 
```

# Prototypical tree shape

#### Bow: the tree is usually tilted to one side and has few branches as it grew. This type of mentortree has single-gender mentees and there are few mentees over different generations. The tree is tiled, thin, and tall.

```python
display(SVG('thin.svg')) 
```

#### S shape: the tree is bended from one side to the other side. Usually, this type of mentor tree has single-gender mentees in the early generations but as it grew, There are more and more opposite gender mentees.

```python
display(SVG('S_shape.svg')) 
```

#### Y-shape: the tree has a large fork at the middle and grew up with canopy in different directions. Usually, this type of mentortree has many mentees in the early generations, while in one generation, some mentees have more mentees than usual in the next couple generations.

```python
display(SVG('Y_shape.svg')) 
```

#### T-shape: a lush branch grew out of the canopy of this tree. Usually, this type of mentortree has many mentees over different generations. And in the most crowded generation, there is one mentee at the middle who became a mentor and have more mentees than others.

```python
display(SVG('T_shape.svg')) 
```

#### Fan: the tree grew with a fan-shaped canopy. This type of mentortree has a growing mentees continously over different generations.

```python
display(SVG('Fan.svg')) 
```
